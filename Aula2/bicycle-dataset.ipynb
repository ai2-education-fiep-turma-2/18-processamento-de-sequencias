# função generator com comentários 
* coluna 14 é a coluna que queremos predizer

```python
def generator(data, min_index, max_index, lookback, delay, batch_size=128, step=6):
    #data array com a sequência temporal
    #cada linha é um item da sequencia e cada coluna é uma característica do item da sequência
    
    #min_index - posição inicial na sequência completa
    #max_index - posiaço final na sequência completa
    #lookback - quantos passos para trás para predizer os próximos itens
    
    # variavei i recebe a posição inicial + qtde de passos para trás
    #É a primeira porção de dados que será observada
    i = min_index + lookback
    #print(min_index, max_index, lookback)
    
    #Laço sem fim (será executado enquanto uma função externa continuar pedindo dados)
    while 1:
        
        # se a posição de i atual for igual ou maior que max_index
        # chegamos no fim da sequência - o índice i volta para o início
        # isso pode ocorrer, caso a rotina continue pedindo registros da sequência, 
        # mesmo tendo chegado ao fim da sequência
        if i + batch_size >= max_index:
            i = min_index + lookback
        
        #print("i ",i)
        
        #variável rows vai receber os índices da sequência principal
        #controla com min para não ultrapassar o limite da sequência
        # tamanho será sempre batch size, ou menor, se batch for maior que o limite
        rows = np.arange(i, min(i + batch_size, max_index))

        #print("shape rows " , rows.shape)
        #atualiza i - variável que controla a posição atual na sequência
        i += len(rows)
            
        # cria samples
        # // divisão inteira - ignora o resto da divisão
        # dimensão do samples
        # quantidade de elementos na sequÊncia x qtde de itens no passado /step x qtde de características
        samples = np.zeros((len(rows),
                           lookback // step,
                           data.shape[-1]))
        
        #cria targets - mesma quantidade de elemtnos de rows (depende do batch size)
        targets = np.zeros((len(rows),))
        
        #loop pela quantidade de linhas do batch
        for j, row in enumerate(rows):
            #índices dos valores na sequência principal
            indices = range(rows[j] - lookback, rows[j], step)
            #print(" indices " , indices, " j ",j)
            
            #atualizar a variável samples de retorno
            samples[j] = data[indices]
            # target é sempre a posição + delay - coluna fixa no target
            # coluna 14 é a coluna que queremos predizer
            targets[j] = data[rows[j] + delay][14]
        yield samples, targets
```

```python
#lookback = 1440
lookback = 7 # timesteps no passado
step = 1      # sample de dados
#delay = 144
delay = 1    # time steps no futuro
#batch_size = 128
batch_size = 10

train_gen = generator(biciclesdata, # dados de entrada
                      min_index=0,
                      max_index=2000,#200000,
                      lookback=lookback,
                      delay=delay,

                      step=step,
                      batch_size=batch_size)
val_gen = generator(biciclesdata,
                    min_index=2000,#200001,
                    max_index=2500,#300000,
                    lookback=lookback,
                    delay=delay,

                    step=step,
                    batch_size=batch_size)
test_gen = generator(biciclesdata,
                     min_index=2500,#300001,
                     max_index=2750,#None,
                     lookback=lookback,
                     delay=delay,

                     step=step,
                     batch_size=batch_size)

val_steps = (2750 - 2500 - lookback)
test_steps = (len(biciclesdata) - 2750 - lookback)
```

```python
def plot_loss(history):
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs = range(1, len(loss) + 1)
    plt.figure()
    plt.plot(epochs, loss, 'bo', label='Training loss')
    plt.plot(epochs, val_loss, 'b', label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()
    plt.show()
```

```python
model = Sequential()
model.add(layers.GRU(32,
                    dropout=0.2,
                    recurrent_dropout=0.2,
                    input_shape=(None, biciclesdata.shape[-1])))
model.add(layers.Dense(1))
model.compile(optimizer=RMSprop(), loss='mae')   
```

```python
epochs=100
steps_p_epochs=100

#steps_p_epochs=500
#epochs=20

history = model.fit_generator(train_gen,
steps_per_epoch=steps_p_epochs, epochs=epochs, validation_data=val_gen, validation_steps=val_steps)
```

```python
plot_loss(history)
```
